```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/datasets_sample/haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('/content/drive/MyDrive/datasets_sample/GettyImages-1190285600.png')
```

```python
faces = face_cascade.detectMultiScale(image= img , scaleFactor= 1.1 , minNeighbors= 5 )
```

```python
for (x, y, w, h) in faces:
  cv2.rectangle(img, (x,y), (x+w , y+h), (255,0,0),2)
```

```python
print(len(faces),'faces detected!')
```

```python
finalimage = cv2.cvtColor(img , cv2.COLOR_RGB2BGR)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(finalimage)
plt.axis('off')
plt.show()
```
